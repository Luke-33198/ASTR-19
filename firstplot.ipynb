```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
```

```python
y = np.exp(x)
```

```python
def awesome_plot(x,y, flag_save=True, xlabel='Time [milliseconds]', ylabel='Awesomeness', lcolor='red', fs=14, fname='firstplot.png'): 
    f, ax = plt.subplots(1,1,figsize=(4,4)) 
    #plot y vs. x 
    ax.plot(x,y,color=lcolor,linewidth=1.5) 
    #label our axes 
    ax.set_xlabel(xlabel,fontsize=fs) 
    ax.set_ylabel(ylabel,fontsize=fs) 
```

```python
awesome_plot(x,y)
```
