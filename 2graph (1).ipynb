```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y1,label='sin graph',linestyle=':')
axes[1].plot(x,y2,label='cos graph',color='orange')
axes[0].legend()
axes[1].legend()
plt.savefig("multi_graph.pdf")
```
