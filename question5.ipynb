```python
import numpy as np
```

```python
from tabulate import tabulate
```

```python
def x(n):
    return np.sin(n)
```

returns a sin value with an input

```python
def y(n):
    return np.cos(n)
```

returns a cos value with an input

```python
a = np.linspace(0,(2*np.pi),1000)
```

assigns the value a with 1000 entries between the values of 0 and 2pi

```python
table_headers = ["x", "y"]
```

assigns table headers

```python
axisx = []
for i in range(10):
    axisx.append(x(a)[i])
```

creates an empty list axisx and assigns the empty list with the first 10 values of sin(a) with a being previously stated

```python
axisy = []
for i in range(10):
    axisy.append(y(a)[i])
```

does the same as above mentions with axisx but creates a new list being axisy and instead of sin its cos(a)

```python
table_data = [(a, b) for a, b in zip(axisx, axisy)]
```

tabulates the data with the lists above

```python
python_table = tabulate(table_data, tablefmt="grid", headers=table_headers,
floatfmt=".3f")
```

the tablulate function with the values previously assigned

print(python_table)

prints the table
